### Correlacion

Como pide el enunciado, haremos una correlacion y un mapa de calor, para comprobar si hay correlacion entre las variables, empezare realizando una prueba de normalidad de las variables, para asi comprobar si usare**pearson** o **spirman**

```python
##Prueba de shapiro Wilks para comprobar normalidad de las variables
```

```python
## Miramos la normalida de las 2 variables
Sepal_Length = df['SepalLengthCm']
Sepal_Width = df['SepalWidthCm']
Petal_Length = df['PetalLengthCm']
Petal_Width = df['PetalWidthCm']
```

```python
# Shapiro-Wilk test variable Sepal_Length
# =================
shapiro_test = stats.shapiro(Sepal_Length)
shapiro_test

```

```python
# Shapiro-Wilk test variable Sepal_Width
# =========================================
shapiro_test = stats.shapiro(Sepal_Width)
shapiro_test
```

```python
# Shapiro-Wilk test variable Petal_Length
# =========================================
shapiro_test = stats.shapiro(Petal_Length)
shapiro_test
```

```python
# Shapiro-Wilk test variable Petal_width
# =========================================
shapiro_test = stats.shapiro(Petal_Width)
shapiro_test
```

Como vemos no todos sigue la normalidad, la unica variale que sigue la normalidad es 
Sepal width, ya que su p-valor es superior a 0.05, graficare, pero cveremos que no siguen la normalidad

```python
# Gráfico Q-Q Sepal_Length
# ==============================================================================
fig, ax = plt.subplots(figsize=(7,4))
sm.qqplot(Sepal_Length,
    fit   = True,
    line  = 'q',
    alpha = 0.4,
    lw    = 2,
    ax    = ax
)
ax.set_title('Gráfico Q-Q delSueldo', fontsize = 10,
             fontweight = "bold")
ax.tick_params(labelsize = 7) ## como podemos ver esta variable  no hay normalidad de los datos, ya que en la cola se va de la linea
```

```python
# Gráfico Q-Q Sepal_Length
# ==============================================================================
fig, ax = plt.subplots(figsize=(7,4))
sm.qqplot(Sepal_Width,
    fit   = True,
    line  = 'q',
    alpha = 0.4,
    lw    = 2,
    ax    = ax
)
ax.set_title('Gráfico Q-Q delSueldo', fontsize = 10,
             fontweight = "bold")
ax.tick_params(labelsize = 7) ## como podemos ver esta variable se asemeja mas a la normalidad, aunque no esten pegados
```

```python
# Gráfico Q-Q Sepal_Length
# ==============================================================================
fig, ax = plt.subplots(figsize=(7,4))
sm.qqplot(Petal_Length,
    fit   = True,
    line  = 'q',
    alpha = 0.4,
    lw    = 2,
    ax    = ax
)
ax.set_title('Gráfico Q-Q delSueldo', fontsize = 10,
             fontweight = "bold")
ax.tick_params(labelsize = 7) ## como podemos verni por asomo sigue la normalidad
```

#### KNN

#### Forma 1

Elijo 10 knn vecino, para empezar a probar la prediccion

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

acc = []

for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train,y_train)
    yhat = knn.predict(X_test)
    acc.append(accuracy_score(y_test,yhat))
    print("For k = ",i," : ",accuracy_score(y_test,yhat))
```

Grafico, pero como puedo ver arriba, todos los vecinos predicen igual

```python
plt.figure(figsize=(8,6))
plt.plot(range(1,10),acc, marker = "o")
plt.xlabel("Value of k")
plt.ylabel("Accuracy Score")
plt.title("Finding the right k")
plt.xticks(range(1,10))
plt.show() ## No merece la pena graficarlo, ya que la prediccion es la misma
```

En el algoritmo de la Maquina de Vector soporte, probare los diferentes kernel, el kernel lineal, polinomial, el rbf y el sigmoide, probare cual de los 4 predice mejor 

### Using Linear Kernel.

```python
# Ajuste de la clasificación SVC al conjunto de entrenamiento con kernel lineal
from sklearn.svm import SVC
svcclassifier = SVC(kernel = 'linear', random_state = 0)
svcclassifier.fit(X_train, y_train)
```

```python
# Predicción de los resultados del conjunto de pruebas
y_pred = svcclassifier.predict(X_test)
print(y_pred)
```

```python
# Veamos el valor real y predicho uno al lado del otro
y_compare = np.vstack((y_test,y_pred)).T
# valor real en el lado izquierdo y valor predicho en el lado derecho
#Imprimimops 5 valores
y_compare[:5,:]
```

```python
#Creamos la matriz de confusion
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
#Calculasmosa la precision a partir de la matriz de confusion
a = cm.shape
corrPred = 0
falsePred = 0

for row in range(a[0]):
    for c in range(a[1]):
        if row == c:
            corrPred +=cm[row,c]
        else:
            falsePred += cm[row,c]
print('Correct predictions: ', corrPred)
print('False predictions', falsePred)
kernelLinearAccuracy = corrPred/(cm.sum())
print ('Accuracy of the SVC Clasification is: ', corrPred/(cm.sum()))
```
